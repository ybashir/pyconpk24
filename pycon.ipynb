```python
from ipyvizzu import Data, Config, Style
from ipyvizzustory import Story, Slide, Step
import pandas as pd
df = pd.read_csv("data/indices_data.csv")
df = df.astype({"Year": str})
langlist = df['Language'].unique().tolist()
data = Data()
data.add_df(df)
colors = {
    
    'all':'#c0c0c0dd',
    'C++':'#4379ABdd',
    'C':'#96CCEBdd',
    'Python':'#3DA443dd',
    'Java':'#FFBC71dd',
    'C/C++':'#B976A3dd',
    'JavaScript':'#DDB3A4dd',
    'SQL':'#BA9900dd',
    'GoLang':'#F7CD4Bdd',
    'Rust':'#77BEB6dd',
    'React':'#c0c0c0dd',    
    'C#':'#F14A54dd',
    'TypeScript':'#FF9797dd',
    'Typescript':'#7B706Edd',
    'AI':'#A3745Cdd',
    'React':'#B976A3dd'
    
    #Default color palette
    #B462ACFF #F492FCFF #9C7EF4FF #9C52B4FF #6CA2FCFF #5C6EBCFF #7C868CFF 
    #AC968CFF #4C7450FF #AC7A4CFF #7CAE54FF #4C7450FF #9C1A6CFF 
    
    #Tableau Color Palette
    #4379AB #96CCEB #FF8900 #FFBC71 #3DA443 #76D472 #BA9900 #F7CD4B
    #249A95 #77BEB6 #F14A54 #FF9797 #7B706E #BCB0AB #E16A96 #FFBCD3
    #B976A3 #DCA3CA #A3745C #DDB3A4
    
}


colorPalette = [colors['all']] * len(langlist)
colorPalette[langlist.index("Python")] = colors['Python']
style = Style({
    "tooltip": {
        "fontSize": "1.4em"
    },
    "title": {
        "paddingTop": "1.2em",
        "fontSize": "2.0em"
    },
    "legend": {
        "label": {
            "fontSize": "2.0em"
        },
        "width": "16em"
    },
    "logo": {
        "width": "6em"
    },
    
    "plot": {
        "paddingLeft":"12em",
        "marker": {
            "label": {
                "fontSize": "2.0em",
                "color":"#888888",
                "maxFractionDigits": 0
            },
            "borderWidth": "4px",
            "colorPalette": ' '.join(colorPalette)
        },
        "yAxis": {
            "label": {
                "fontSize": "2.0em",
            },
            "title": {
                "color": "#ffffff00"
            },
            "interlacing": {
                "color": "#ffffff00"
            },
        },
        "xAxis": {
            "label": {
                "fontSize": "2.0em",
                "paddingTop": "1em",
            },
            "title": {
                "fontSize": "2.0em",
                "paddingTop": "2.0em"
            },
        },
    },
})



# Add the slide to the story
popularity_indices = Slide(
    Step(
        Config({
            

            "channels":{
                "x":None,
                "y":None,
                "color":"Index",
                "label":"Index",
                "size":"Index",
                
                
              },
            "split":True,
            "title":"Popularity Indices"
        }), Style({
            "plot": {
                "marker": {
                    "label":{
                        "fontSize":"2em",
                        "color":"#ffffffff"
                    },
                    "colorPalette": '#4171CDFF #03AE71FF\
                                     #F4941BFF #F4C204FF'
                }
            }
        })
    )
)
python_ranking_medals = Slide(
    Step(
        Data.filter("""(record.Year === '2020' || \
                        record.Year === '2021' || \
                        record.Year === '2023') && \
                        (record.Language === 'Python')"""),
        Config({
            

            "channels":{
                "x":["Index"],
                "y":{
                    "set":["Year"],
                    
                },
                "color":"Rank",
                "label":"Rank",                
                
              },
            "split":True,
            "title":"Python's Ranking Across Popularity Indices"
        }), Style({
            "plot": {
                "marker": {
                    "label":{
                        "fontSize":"2em",
                        "color":"#ffffffff"
                    },
                    "colorPalette": '#ffd700',
                    "colorGradient": '#F1CE63FF 0.000000,\
                                        #F1CE63FF 0.250000, \
                                        #BAB0ACFF 0.500000, \
                                        #9D7660ff 0.750000, \
                                        #E15759FF 1.000000'
                },
                "xAxis":{"label":{"fontSize":"2em"}},
                "yAxis":{"label":{"fontSize":"2em"}},
                
            },
        })
    )
)


python_TIOBE_2017 = Slide(
    Step(Data.filter("""
        
        record.Index === 'TIOBE' 
        && record.Year === '2012'
        && record.Language !== 'C/C++' 
        && record.Rank < 12.0

        
        """),
         Config({
             "x": "Ratings",
             "y": "Language",
             "sort": "byValue",
             "color": "Language",
            
             "title": "Python's TIOBE Rank in 2012: 7",
             "geometry":"rectangle",
         }),
         style,
         duration=0))

python_TIOBE_race = Slide()
for i in range(2012, 2025):
    rank = int(df[(df['Language']=='Python')\
                  &(df['Index']=='TIOBE')\
                  &(df['Year']==str(i))]['Rank'].item())
    filters = """
            record.Index === 'TIOBE' 
            && record.Year === '{}' 
            && record.Language !== 'C/C++' 
            && record.Rank < 12.0
    """.format(i)

    step = Step(
        Data.filter(filters),
        Config({
            "x": "Ratings",
            "y": "Language",
            "color": "Language",
            "sort": "byValue",
            "title": "Python's {} TIOBE Ranking: {}".format(i,rank)
        }),
        duration=1,
        x={
            "easing": "linear",
            "delay": 0
        },
        y={"delay": 0},
        show={"delay": 0},
        hide={"delay": 0},
        title={
            "duration": 0,
            "delay": 0
        },
    )
    python_TIOBE_race.add_step(step)

cp = colorPalette.copy()
cp[langlist.index("C")] = colors['C']
cp[langlist.index("C++")] = colors['C++']
cp[langlist.index("C/C++")] = colors['C/C++']
cp[langlist.index('Java')] = colors['Java']
cp[langlist.index('SQL')] = colors['SQL']
cp[langlist.index('JavaScript')] = colors['JavaScript']
cp[langlist.index('GoLang')] = colors['GoLang']
cp[langlist.index('Rust')] = colors['Rust']
cp[langlist.index('React')] = colors['React']
cp[langlist.index('C#')] = colors['C#']
cp[langlist.index('TypeScript')] = colors['TypeScript']
cp[langlist.index('AI')] = colors['AI']
cp[langlist.index('React')] = colors['React']


python_TIOBE_linechart = Slide()


langs = df[(df['Year']=='2024')&(df['Index']=='TIOBE')&(df['Rank']<7)]\
        .sort_values(by=["Rank"])['Language'].tolist()
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs if x != 'C/C++'])

python_TIOBE_linechart.add_step(
    Step(
        Data.filter("record.Index === 'TIOBE' && ({})".format(filterby)),
        Config({
            "label":None,
            "x": "Year",
            "y": ["Ratings"],
            "geometry": "line",
            "sort": "none",
            "color": "Language",
            "title": "Python in the TIOBE Index",
            "split": False
        }), Style({
            "plot": {
                "marker": {
                    "colorPalette": ' '.join(cp)
                }
            },
        })
    
    ))


langs = df[(df['Year']=='2024')&(df['Index']=='TIOBE')&(df['Rank']<8)]['Language'].tolist()
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs if x not in ['C','C++']])

python_TIOBE_with_c = Slide(
    Step(
        Data.filter("record.Index === 'TIOBE' && ({})".format(filterby)),
        Config({          
            "title": "Ranking in the TIOBE Index with C/C++ Combined"
        })
    ))




langs = df[(df['Year']=='2024')&(df['Index']=='PYPL')&(df['Rank']<6)]['Language'].tolist()
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs])
pypl = Slide(
    Step(
        Data.filter("record.Index === 'PYPL' && ({})".format(filterby)),
        Config({

            "title": "PYPL"
        })
    )
)



langs = df[(df['Year']=='2023')&(df['Index']=='StackOverflow')&(df['Rank']<8)]['Language'].tolist()
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs \
                        
                        if x not in ['HTML/CSS','Bash/Shell (all shells)']])


stackoverflow = Slide(
    Step(
        Data.filter("record.Index === 'StackOverflow' && ({})".format(filterby)),
        Config({            
            "channels": {
                    "x": { "range":{"min":1,"max":7}},                    
                },
            "title": "StackOverflow",
            
        })
    )
)




langs = df[(df['Year']=='2023')&(df['Index']=='HackerNews')&(df['Rank']<8)]['Language'].tolist()
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs if x not in \
        ['HTML/CSS','AWS','Machine Learning','DevOps','Kubernetes','Data Science','Docker','Django']])

hackernews = Slide(
    Step(
        Data.filter("record.Index === 'HackerNews' && ({})".format(filterby)),
        Config({
            "title": "HackerNews Who's Hiring Job Count",    
              "channels": {
                    "x": { "range":{"min":1,"max":'auto'}},                    
                },
        })
    )
)
# Create story object, add data to it
story = Story(data=data, style=style)
story.set_size(width="100%", aspect_ratio=16 / 8.5)
story.add_slide(popularity_indices)
story.add_slide(python_ranking_medals)
story.add_slide(python_TIOBE_2017)
story.add_slide(python_TIOBE_race)
story.add_slide(python_TIOBE_linechart)
story.add_slide(python_TIOBE_with_c)
story.add_slide(pypl)
story.add_slide(stackoverflow)
story.add_slide(hackernews)
story.set_feature("tooltip",False)
story.play()
```

```python
import pandas as pd
from ipyvizzu import Data, Config, Style
from ipyvizzustory import Story, Slide, Step

df = pd.read_csv("data/linkedin.csv")

data = Data()
data.add_df(df)




colors="""
    #4379AB #96CCEB #FF8900 #FFBC71 #3DA443 #76D472 #BA9900 #F7CD4B
    #249A95 #77BEB6 #F14A54 #FF9797 #7B706E #BCB0AB #E16A96 #FFBCD3
    #B976A3 #DCA3CA #A3745C #DDB3A4
    """
colors=" ".join(str.split(colors)) 
style = Style(
    {
         "plot":{"paddingLeft":"15em","marker":{
             "label":{"fontSize":"2em","fontWeight":"normal","color":"#ffffffff"},
             "colorPalette":colors}},
          "legend": {
        "label": {
            "fontSize": "2em",
            "maxFractionDigits": 0
        },
        "width": "14em"
        },
        "caption":{
            "fontSize":"1.5em"
        }
    }
)
story = Story(data=data,style=style)
story.set_size(width="100%", aspect_ratio=16 / 8.5)


total_python_developers=Slide(
    Step(
        Data.filter("record.Dimension==='Location' && record.Language==='Python' && record.Country==='Pakistan'"),
        Config({
             "channels": {
                "size": {"set": ["Total"]},
                "label": {"set": ["Total"]},
                 "color":{"set":["Country"]}
            },
            "caption":"Source: LinkedIN",
            "legend": None,
            "geometry": 'circle',
            "title": "Total Python Developers in Pakistan",
            
        })
    )
)

compared_with_india =Slide(
    Step(Data.filter("""record.Dimension==='Location' && record.Language === 'Python' && \
    
        (record.Country==='Pakistan'||record.Country=='India'|| record.Country==='USA') || record.Country==='Ukraine'"""),
         Config(
            {"title":"Small compared to India and USA"}),
         Style({"plot":{"marker":{"label":{"color":"#ffffffaa"}}}})
        )
)
#Location Distribution

python_vs_others = Slide(
    Step(
        Data.filter("record.Country=='Pakistan' && record.Dimension==='Location' && record.Language!='Others'"),
        Config({
            "channels": {
                "size": {"set": ["Total"]},
                "color": {"set": ["Language"]},
                "label": {"set": ["Language","Total"]}
            },
            "legend": "color",
            "geometry": 'rectangle',
            "title": "Python vs other languages in Pakistan",
        }),
        Style({
            "plot":{"marker":{"label":{"fontSize":"2.5em","color":"#ffffffff"},"colorPalette":colors},
                    
            },
            
            
        })
      
    )
)

python_devs_in_pakistan = Slide(
    Step(
        Data.filter("record.Country=='Pakistan' && record.Dimension==='Location' && record.Language=='Python'"),
        Config({
            "title": "Total Python Developers in Pakistan",
        }),
         Style({"plot":{"marker":{"label":{ "fontSize":"2em", "color":"#ffffffff"}}}})
    )
)


distribution_across_provinces = Slide(
    Step(
        Data.filter("record.Country=='Pakistan' && record.Dimension==='Location' && record.Language=='Python'"),
        Config({
            "geometry": "rectangle",
            "channels": {
                "y": {"set":"Province","range":{"min":"10%"}},
                "x": {"set":["Total",]},
                "color": "Province",
                "label": {"set": "Total"
                }
            },
            "legend": "auto",
            "title": "Distribution across provinces",
            "sort":'byValue',            
      
          
        }),
        Style({
            "plot":{
                "marker":{
                    "label":{
                            "fontSize":"2em",
                        "color":"#888888"
                    }
                },
                "paddingLeft":"10em",
   
                "xAxis":{"label":{"fontSize":"1.5em"}},
                "yAxis":{"label":{"fontSize":"1.5em",}},
                
            }
        })
    )
)

distribution_across_cities = Slide(
    Step(
        Data.filter("record.Country=='Pakistan' && record.Dimension==='Location' \
        && record.Language=='Python' && \
        (record.City=='Lahore' || record.City=='Karachi' || record.City == 'Islamabad' \
        || record.City == 'Multan' || record.City=='Rawalpindi' || record.City=='Hyderabad')"),
        Config({
            "channels": {
                "x": {
                    "attach": ["City"],
                },

                "color":{
                    #"detach":"Province",
                    "set":"City"
                },
                "label":None,
                "size":["Total"]
                
            },
            "split":"true",
            "legend": "color",
            "title": "Distribution across cities",
            
        }),
         Style({"plot":{"marker":{"label":{"fontSize":"2em","fontWeight":"normal","color":"#ffffffff"}}}})
    ))






#Experience Distribution

experience_distribution = Slide(
    
    Step(
        Data.filter("record.Country=='Pakistan' && record.Dimension==='Experience' && record.Language=='Python'"),
        Config({
            "channels": {
                "y": {
                    "set":None
                },
                "x":{
                    "set":"Country"
                },
                "color":{
                    "set":"Language"
                },
                "label":{
                    "set":"Total",
                    
                },
                "size":{
                    "set":"Total",
                    
                },
                
                
            },
            
            "legend": None,
            "title": "Distribution by Experience",
            "reverse":"true",
            
            
            
        }),
        duration=0
    ))

python_experience_distribution = Slide(
    
    Step(
        Data.filter("record.Country=='Pakistan' && record.Dimension==='Experience' && record.Language=='Python'"),
        Config({
            "channels": {
                "y": {
                    "set":"Total"
                },
                "x":{
                    "set":"Experience"
                },
                "color":{
                    "set":"Experience"
                },
                "label":{
                    "set":"Total",
                    
                },
                "size":{
                    "set":"Total",
                    
                },
                
                
            },
            "sort":"none",
            "legend": "color",
            "title": "Python distribution by experience",
            
        }),
    Style(
            {
                 "plot":{"paddingLeft":"15em","marker":{
                     "label":{"fontSize":"2em","fontWeight":"normal","color":"#888888","maxFractionDigits":0},
                     "colorPalette":colors}},
            }
        ),
    ))

java_experience_distribution = Slide(
    
    Step(
        Data.filter("record.Country=='Pakistan' && record.Dimension==='Experience' && record.Language=='Java'"),
        Config({
            "channels": {
                "y": {
                    "set":"Total"
                },
                "x":{
                    "set":"Experience"
                },
                "color":{
                    "set":"Experience"
                },
                "label":{
                    "set":"Total",
                    
                },
                "size":{
                    "set":"Total",
                    
                },
                
                
            },
            "sort":"none",
            "legend": "color",
            "title": "Java distribution by experience",
            
        }),
    Style(
            {
                 "plot":{"paddingLeft":"15em",
                         "marker":{"label":{"fontSize":"1.5em","fontWeight":"normal","color":"#888888"},
                                   "colorPalette":colors}},
            }
        ),
    ))


university_distribution = Slide(
    
    Step(
        Data.filter("record.Country=='Pakistan' && record.Dimension==='University' && record.Language=='Python'"),
        Config({
            "channels": {
                "y": {
                    "set":["University"],
               
                },
                "x":{
                    "set":["Total",]
                },
                "color":{
                    "set":"University"
                },
                "label":{
                    "set":"Total",
                    
                },
                "size":{
                    "set":"Total",
                    
                },
                
                
            },
            
            "legend": "null",
            "title": "Distribution by University",
            "sort":"byValue",
            "reverse":"false"
         
        }),
        Style(
            {
                 "plot":{"paddingLeft":"15em","marker":{"label":{"fontSize":"1.5em","fontWeight":"normal","color":"#888888"},"colorPalette":colors}},
            }
        ),
        duration=0
    ))

story.add_slide(total_python_developers)
story.add_slide(compared_with_india)
story.add_slide(python_vs_others)
#story.add_slide(python_devs_in_pakistan)

story.add_slide(distribution_across_provinces)
story.add_slide(distribution_across_cities)

story.add_slide(experience_distribution)
story.add_slide(python_experience_distribution)
story.add_slide(java_experience_distribution)


story.add_slide(university_distribution)

story.set_feature("tooltip",True)
story.play()
```
