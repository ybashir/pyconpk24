```python
import pandas as pd
from datetime import datetime
from dateutil import parser
def parse_date(date_string):
    date_part = ' '.join(date_string.split(' ')[:4])  # Extracts "Thu Jul 01 2004"
    parsed_date = datetime.strptime(date_part, "%a %b %d %Y").date()    
    return parsed_date

df = pd.read_csv("data/tiobe.csv")
df['C/C++'] = df['C']+df['C++']
df = pd.melt(df,id_vars=df.columns[:1],value_vars=df.columns[1:],var_name="Language",value_name="Ratings" )
df["Date"] = pd.to_datetime(df["Date"])
df = df[df['Date'].dt.month == 2]
df["Index"] = "TIOBE"
df['Year'] = df['Date'].dt.year 


df1= pd.read_csv("data/pypl.csv")
df1= pd.melt(df1,id_vars=df1.columns[:1],value_vars=df1.columns[1:],var_name="Language",value_name="Ratings" )
df1['Date'] = df1['Date'].apply(parse_date)
df1["Date"] = pd.to_datetime(df1["Date"])
df1 = df1[df1['Date'].dt.month == 2]
df1["Index"] = "PYPL"
df1['Ratings'] = df1['Ratings']*100
df1['Year'] = df1['Date'].dt.year 

df2= pd.read_csv("data/stackoverflow.csv")
df2= pd.melt(df2,id_vars=df2.columns[:1],value_vars=df2.columns[1:],var_name="Language",value_name="Ratings" )
df2["Index"] = "StackOverflow"
df2["Ratings"] = df2["Ratings"].fillna(0)

df3= pd.read_csv("data/hackernews.csv")

df = pd.concat([df1,df,df2,df3])
df = df[df['Year']>2010]
df['Year'] = df['Year'].apply(str)

df['Rank'] = df.groupby(['Year','Index'])['Ratings'].rank(ascending=False,method='dense')
df['Rank'] =  df['Rank'].fillna(1000)
langlist = df['Language'].unique().tolist()

print(df2)
print(df3)
```

```python
from ipyvizzu import Data, Config, Style
from ipyvizzustory import Story, Slide, Step

data = Data()
data.add_df(df)
colors = {
    'all':'#c0c0c0',
    'C++':'#98fb98',
    'C':'#db7093',
    'Python':'#6b8e23',
    'Java':'#48d1cc',
    'C/C++':'#ff4500',
    'JavaScript':'#6a5acd',
    'SQL':'#fa8072',
    'GoLang':'#ffd700',
    'Rust':'#ff69b4'
    
}


colorPalette = [colors['all']] * len(langlist)
colorPalette[langlist.index("Python")] = colors['Python']
style = Style({
    "tooltip": {
        "fontSize": "1.2em"
    },
    "title": {
        "paddingTop": "1.2em",
        "fontSize": "2.5em"
    },
    "legend": {
        "label": {
            "fontSize": "1.2em"
        },
        "width": "16em"
    },
    "logo": {
        "width": "6em"
    },
    "plot": {
        "marker": {
            "label": {
                "fontSize": "1.5em"
            },
            "borderWidth": "4px",
            "colorPalette": ' '.join(colorPalette)
        },
        "yAxis": {
            "label": {
                "fontSize": "1.2em",
            },
            "title": {
                "color": "#ffffff00"
            },
            "interlacing": {
                "color": "#ffffff00"
            },
        },
        "xAxis": {
            "label": {
                "fontSize": "1.4em",
                "paddingTop": "1em",
            },
            "title": {
                "fontSize": "1.4em",
                "paddingTop": "2.5em"
            },
        },
    },
})
# Create story object, add data to it
story = Story(data=data, style=style)
story.set_size(width="100%", aspect_ratio=16 / 8)

preslide = Slide(
    Step(Data.filter("""
        
        record.Index === 'TIOBE' 
        && record.Year === '2012'
        && record.Language !== 'C/C++' 
        && record.Rank < 12.0

        
        """),
         Config({
             "x": "Ratings",
             "y": "Language",
             "sort": "byValue",
             "color": "Language",
             "title": "Python's TIOBE Rank in 2012: 7"
         }),
         duration=1))

slide0 = Slide()
for i in range(2012, 2025):
    rank = int(df[(df['Language']=='Python')&(df['Index']=='TIOBE')&(df['Year']==str(i))]['Rank'].item())-1
    filters = """
            record.Index === 'TIOBE' 
            && record.Year === '{}' 
            && record.Language !== 'C/C++' 
            && record.Rank < 12.0
    """.format(i)

    step = Step(
        Data.filter(filters),
        Config({
            "x": "Ratings",
            "y": "Language",
            "color": "Language",
            "sort": "byValue",
            "title": "Python's TIOBE Rank in {}: {}".format(i,rank)
        }),
        duration=1,
        x={
            "easing": "linear",
            "delay": 0
        },
        y={"delay": 0},
        show={"delay": 0},
        hide={"delay": 0},
        title={
            "duration": 0,
            "delay": 0
        },
    )
    slide0.add_step(step)

slide1 = Slide()
cp = colorPalette.copy()

cp[langlist.index("C")] = colors['C']
cp[langlist.index("C++")] = colors['C++']
langs = df[(df['Year']=='2024')&(df['Index']=='TIOBE') ] \
            .sort_values(by="Ratings",ascending=False)['Language'] \
            .tolist()[:11]
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs if x != 'C/C++'])

slide1.add_step(
    Step(
        Data.filter("""
        record.Index === 'TIOBE' &&
        ({})
        """.format(filterby)),
        Config({
            "x": "Year",
            "y": "Ratings",
            "geometry": "line",
            "sort": "none",
            "color": "Language",
            "title": "Python in the TIOBE Index"
        }), Style({
            "plot": {
                "marker": {
                    "colorPalette": ' '.join(cp)
                }
            },
        })))

cp = colorPalette.copy()
cp[langlist.index("C/C++")] = colors['C/C++']
langs = df[(df['Year']=='2024')&(df['Index']=='TIOBE') ] \
            .sort_values(by="Ratings",ascending=False)['Language'] \
            .tolist()[:12]
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs if x not in ['C','C++']])

slide2 = Slide(
    Step(
        Data.filter("""
        record.Index === 'TIOBE' &&
        ({})
        """.format(filterby)),
        Config({
            "x": "Year",
            "y": "Ratings",
            "geometry": "line",
            "color": "Language",
            "split": False,
            "title": "Second in the TIOBE Index with C/C++ Combined"
        })))

slide2.add_step(
    Step(Style({
        "plot": {
            "marker": {
                "colorPalette": ' '.join(cp)
            }
        },
    }, )))

cp = colorPalette.copy()

cp[langlist.index('Java')] = colors['Java']
cp[langlist.index('JavaScript')] = colors['JavaScript']
langs = df[(df['Year']=='2024')&(df['Index']=='PYPL') ] \
            .sort_values(by="Ratings",ascending=False)['Language'] \
            .tolist()[:10]
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs])
slide3 = Slide(
    Step(
        Data.filter("""
        record.Index === 'PYPL' &&
        ({})
        """.format(filterby)),
        Config({
            "x": "Year",
            "y": "Ratings",
            "geometry": "line",
            "color": "Language",
            "sort": "none",
            "title": "First in the PYPL Index"
        }), Style({
            "plot": {
                "marker": {
                    "colorPalette": ' '.join(cp)
                }
            },
        })))

cp = colorPalette.copy()

cp[langlist.index('SQL')] = colors['SQL']
cp[langlist.index('JavaScript')] = colors['JavaScript']
langs = df[(df['Year']=='2023')&(df['Index']=='StackOverflow') ] \
            .sort_values(by="Ratings",ascending=False)['Language'] \
            .tolist()[:12]
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs \
                        
                        if x not in ['HTML/CSS','Bash/Shell (all shells)']])
slide4 = Slide(
    Step(
        Data.filter("""
        record.Index === 'StackOverflow' &&
        ({})
        """.format(filterby)),
        Config({

            "geometry": "line",
            
            "sort": "none",
            "channels": {
                    "y": {
                        "set": ["Ratings"],
                    },
                    "x": {
                        "set": ["Year"],
                        "range":{
                                "min":1
                        }
                    },
                    "color": "Language",
                },
            "title": "Second in StackOverflow's 'Languages Worked With'",
            
        }), Style({
            "plot": {
                "marker": {
                    "colorPalette": ' '.join(cp)
                }
            },
        })))


cp = colorPalette.copy()

cp[langlist.index('GoLang')] = colors['GoLang']
cp[langlist.index('Rust')] = colors['Rust']
langs = df[(df['Year']=='2024')&(df['Index']=='HackerNews') ] \
            .sort_values(by="Ratings",ascending=False)['Language'] \
            .tolist()[:13]
filterby = ' || '.join(["record.Language === '{}'".format(x) for x in langs if x not in \
    ['HTML/CSS','Bash/Shell (all shells)','DevOps','Typescript','Kubernetes','Data Science','Docker','Django']])
slide5 = Slide(
    Step(
        Data.filter("""
        record.Index === 'HackerNews' &&
        ({})
        """.format(filterby)),
        Config({

            "geometry": "line",
            "sort": "none",
            "channels": {
                    "y": {
                        "set": ["Ratings"],
                    },
                    "x": {
                        "set": ["Year"],
                      
                    },
                    "color": "Language",
                },
            "title": "HackerNews Who's Hiring Job Count",
            
        }), Style({
            "plot": {
                "marker": {
                    "colorPalette": ' '.join(cp)
                }
            },
        })))

# Add the slide to the story
story.add_slide(preslide)
story.add_slide(slide0)
story.add_slide(slide1)
story.add_slide(slide2)
story.add_slide(slide3)
story.add_slide(slide4)
story.add_slide(slide5)
story.set_feature('tooltip',True)
story.play()
```
